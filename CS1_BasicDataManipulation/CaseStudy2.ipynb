Write a function to label an incident that was both successful and suicidal
Hint: We can use apply to use a function column wise where you can use columns success and suicide in the data.

```python
import numpy as np
condition=(tt['suicide']==1)&(tt['success']==1)

tt['SS']=np.where(condition,1,0)
```

```python
def fc(row):
    if row['country_txt']in['India','Pakistan','Afghanistan']:
        return 'Local'
    else:
        return 'ROW'

tt['Local']=tt.apply(fc,axis=1)
```
